```python
import os
import glob
import numpy as np
import pandas as pd
```

## Extracting the sensor data of each candidate

```python
#TIME_LENGTH = 12
PATH = './drive/MyDrive/SisFall_Preprocessed/'
user_details = pd.read_csv(f'{PATH}/user_details.csv')
os.mkdir('./SisFall_Preprocessed')
for uid in user_details.user_id.values:
    os.mkdir(f'./SisFall_Preprocessed/{uid}')
```

```python
from tqdm.notebook import tqdm

for uid in tqdm(user_details.user_id.values):
    for file_name in tqdm(os.listdir(f'./SisFall_dataset/{uid}')):
        if 'txt' in file_name:
          file = open(f'./SisFall_dataset/{uid}/{file_name}', 'r')
          file = file.read().replace('\n', '').split(';')
          for i in range(len(file)):
              file[i] = file[i].lstrip().replace(' ','').split(',')
              file[i] = [int(x.lstrip().rstrip()) for x in file[i] if x != '']
          file = np.array(file[:-1])
          np.savetxt(f'./SisFall_Preprocessed/{uid}/{file_name}', file)
```

```python
global_df = pd.DataFrame()
files = []
target = pd.read_csv(os.path.join(PATH,'target_values.csv'))
for f in tqdm(glob.glob('./SisFall_Preprocessed/*/*')):
  if 'txt' in f:
    file_name = f.split('/')[-1]
    files.append(file_name)
    df = pd.read_csv(f, sep=' ', header=None)
    df = pd.DataFrame(np.expand_dims(df.T.values,1).tolist()).T
    global_df = global_df.append(df, ignore_index=True)

global_df.columns = [f'C{i}' for i in range(9)]
global_df['filename'] = files
global_df = pd.merge(global_df, target, on='filename', how='left')
global_df.to_csv(os.path.join(PATH,'sensor_data.csv'), index=False)
```

# Extracting target values

```python
import os
import pandas as pd
import numpy as np

files = os.listdir('./SisFall_dataset')
res = []
for f in files:
    if f.startswith('S'):
        sub_files = np.array(os.listdir(f'./SisFall_dataset/{f}'))
        for s in sub_files:
            if s.startswith('D'):
                res.append([s,0])
            else:
                res.append([s,1])
df = pd.DataFrame(res, columns=['filename','target'])
df.to_csv('./drive/MyDrive/SisFall_Preprocessed/target_values.csv', index=False)
```
